```python
def num_col_analysis(col):
    print(df[col].describe())
    
    fig, axes = plt.subplots(ncols=2, figsize=(12, 4))
    
    sns.histplot(data=df[col], kde=True, color='skyblue', ax=axes[0])
    axes[0].set_title("Skew = {}".format(df[col].skew()))
    
    sns.boxplot(data=df, x=col, ax=axes[1])
    axes[1].set_title("Box plot")
    
    plt.show()
```

### How does 5g, nfc, ir_blaster phones affects price

```python
fig, axes = plt.subplots(ncols=3, figsize=(12, 4))

sns.barplot(data=temp_df,x='has_5g',y='price',estimator=np.median, ax=axes[0])
axes[0].set_title("5G vs Price")

sns.pointplot(data=temp_df,x='has_nfc',y='price',estimator=np.median, ax=axes[1])
axes[1].set_title("NFC vs Price")

sns.barplot(data=temp_df,x='has_ir_blaster',y='price',estimator=np.median, ax=axes[2])
axes[2].set_title("IR Blaster vs Price")

plt.show()
```
