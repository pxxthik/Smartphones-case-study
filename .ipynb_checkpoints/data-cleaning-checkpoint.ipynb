## Data Assessing

### Quality Issues

1. **model** some brands written differently like OPPO #19, Oppo #13 and itel #401, Itel #483 and POCO #516, Poco #554 `consistency`

2. **price** unneccesary '₹' and ',' `validity`

3. **price** #610 price 99 `accuracy`

4. **rating** missing values `completeness`

5. **sim** #756 looks like it is not a smartphone `validity`

6. **sim**  has some incorrect values in #no 378, 475, 534, 553, 575, 642, 647, 649, 659, 667, 701, 750, 759, 819, 859, 883, 884, 919, 927, 929, 932, 1002 `validity`

7. **processor** - has some incorrect values for some samsung phones(row # -642, 647, 649, 659, 667, 701, 750, 759, 819, 859, 883, 884, 919, 927, 929, 932, 1002) `validity`

8. **ram** some values are in MB `accuracy`

9. **ram** - incorrect values in rows (441, 485, 534, 553, 584, 610, 613, 642, 647, 649, 659, 667, 701, 750, 759, 819, 859, 884, 919, 927, 929, 932, 990, 1002) `validity`

10. **battery** - incorrect values in rows(113, 151, 309, 365, 378, 441, 450, 553, 584, 610, 613, 630, 642, 647, 649, 659, 667, 701, 750, 756, 759, 764, 819, 855, 859, 884, 915, 916, 927, 929, 932, 990, 1002) `validity`

11. **display** - sometimes frequency is not available `completeness`

12. **display** - incorrect values in rows(378, 441, 450, 553, 584, 610, 613, 630, 642, 647, 649, 659, 667, 701, 750, 759, 764, 819, 859, 884, 915, 916, 927, 929, 932, 990, 1002) `validity`

13. certain phones are foldable and the info is scattered `validity`

14. **camera** - words like Dual, Triple and Quad are used to represent number of cameras and front and rear cameras are separated by '&' `validity`

15. camera - problem with rows (100, 113, 151, 157, 161, 238, 273, 308, 309, 323, 324, 365, 367, 378, 394, 441, 450, 484, 506, 534, 553, 571, 572, 575, 584, 610, 613, 615, 630, 642, 647, 649, 659, 667, 684, 687, 705, 711, 723, 728, 750, 756, 759, 764, 792, 819, 846, 854, 855, 858, 883, 884, 896, 915, 916, 927, 929, 932, 945, 956, 990, 995, 1002, 1016 ) `validity`

16. **card** - sometimes contains info about os and camera `validity`

17. **os** - sometimes contains info about bluetooth and fm radio `validity`

18. **os** - issue with rows (324, 378) `validity`

19. **os** - sometimes contains os version name like lollipop `consistency`

20. missing values in camera, card and os `completeness`

21. datatype of price and rating is incorrect `validity`

### Tidiness Issues

1. **sim** can be split into 3 cols has_5g, has_NFC, has_IR_Blaster

2. **processor** - can be split into processor name, cores and cpu speed

3. **ram** - can be split into 2 cols RAM and ROM

4. **battery** - can be split into battery capacity, fast_charging_available

5. **display** - can be split into size, resolution_width, resolution_height and frequency

6. **camera** - can be split into front and rear camera

7. **card** - can be split into supported, extended_upto

### battery_capacity, fast_charging columns

```python
battery_capacity = df['battery'].str.strip().str.split('with').str.get(0).str.strip().str.findall(r'\b(\d+)\b').str.get(0).astype(float)
```

```python
df1.insert(16, 'battery_capacity', battery_capacity)
```

```python
fast_charging = df1['battery'].str.strip().str.split('with').str.get(1).str.strip().str.findall(r'\d{2,3}')
```

```python
df1.insert(17, 'fast_charging', fast_charging)
```

```python
def fast_charger_extracter(item):
    if type(item) == list:
        if len(item) == 1:
            return item[0]
        return 0
    return -1
```

```python
df1['fast_charging'] = df1['fast_charging'].apply(fast_charger_extracter).astype(int)
```

### num_rear_cameras, num_front_cameras columns

```python
def camera_extractor(text):

  if 'Quad' in text:
    return '4'
  elif 'Triple' in text:
    return '3'
  elif 'Dual' in text:
    return '2'
  elif 'Missing' in text:
    return 'Missing'
  else:
    return '1'
```

```python
num_rear_cameras = df1['camera'].str.strip().str.split('&').str.get(0).apply(camera_extractor)
```

```python
df1.insert(25, 'num_rear_cameras', num_rear_cameras)
```

```python
num_front_cameras = df1['camera'].str.strip().str.split('&').str.get(1).str.strip().fillna('Missing').apply(camera_extractor)
```

```python
df1.insert(26,'num_front_cameras',num_front_cameras)
```

```python
def os_extractor(text):

  if 'Android' in text:
    return 'android'
  elif 'iOS' in text:
    return 'ios'
  elif 'Not Specified':
    return text
  elif 'Harmony' in text or 'Hongmeng' in text or 'EMUI' in text:
    return 'other'
```

```python
df1['os'] = df1['os'].apply(os_extractor)
```
